```python
import pandas as pd
import matplotlib.pyplot as plt

# =====================================================
# ĐỌC DỮ LIỆU
# =====================================================

url = "https://raw.githubusercontent.com/huynhhoc/DataAnalystDeepLearning/main/Data/dulieuxettuyendaihoc.csv"

df = pd.read_csv(url)

# =====================================================
# TẠO CỘT XẾP LOẠI XL1 XL2 XL3
# =====================================================

# =====================================================
# TẠO ĐIỂM TRUNG BÌNH
# =====================================================

df['TBM1'] = (
    df['T1'] +
    df['L1'] +
    df['H1'] +
    df['S1'] +
    df['V1']
) / 5

df['TBM2'] = (
    df['T2'] +
    df['L2'] +
    df['H2'] +
    df['S2'] +
    df['V2']
) / 5

# Nếu không có TBM3 thì tạo giả lập
df['TBM3'] = (
    df['TBM1'] + df['TBM2']
) / 2

# =====================================================
# TẠO XL1 XL2 XL3
# =====================================================

def xep_loai(diem):

    if diem >= 8:
        return 'G'

    elif diem >= 6.5:
        return 'K'

    elif diem >= 5:
        return 'TB'

    elif diem >= 3.5:
        return 'Y'

    else:
        return 'Kém'


df['XL1'] = df['TBM1'].apply(xep_loai)
df['XL2'] = df['TBM2'].apply(xep_loai)
df['XL3'] = df['TBM3'].apply(xep_loai)

# =====================================================
# TẠO CỘT KQXT (ĐẬU / RỚT)
# =====================================================

def ket_qua(row):

    tong_diem = row['DH1'] + row['DH2'] + row['DH3']

    if tong_diem >= 15:
        return 'Đậu'
    else:
        return 'Rớt'


df['KQXT'] = df.apply(ket_qua, axis=1)

# =====================================================
# CÂU 1
# Trực quan học sinh nữ theo XL1 XL2 XL3
# =====================================================

print("===== CÂU 1 =====")

# Lọc học sinh nữ
nu_df = df[df['GT'] == 'F']

# Gom dữ liệu
xl_data = {}

for xl in ['XL1', 'XL2', 'XL3']:

    xl_data[xl] = nu_df[xl].value_counts()

# Tạo DataFrame
xl_df = pd.DataFrame(xl_data).fillna(0)

print(xl_df)

# Biểu đồ cột
xl_df.T.plot(
    kind='bar',
    stacked=False,
    figsize=(10,6)
)

plt.title('Xếp loại học sinh nữ')
plt.xlabel('XL1 - XL2 - XL3')
plt.ylabel('Số lượng học sinh')

plt.grid(axis='y')

plt.show()

# =====================================================
# CÂU 2
# KQXT trên học sinh khối A A1 B thuộc KV1 KV2
# =====================================================

print("===== CÂU 2 =====")

loc_df = df[
    (df['KT'].isin(['A', 'A1', 'B'])) &
    (df['KV'].isin(['1', '2']))
]

kqxt_df = pd.crosstab(
    loc_df['KT'],
    loc_df['KQXT']
)

print(kqxt_df)

kqxt_df.plot(
    kind='bar',
    figsize=(8,5)
)

plt.title('KQXT theo khối thi')
plt.xlabel('Khối thi')
plt.ylabel('Số lượng')

plt.grid(axis='y')

plt.show()

# =====================================================
# CÂU 3
# Số lượng thí sinh từng khu vực theo khối thi
# =====================================================

print("===== CÂU 3 =====")

kv_kt = pd.crosstab(
    df['KT'],
    df['KV']
)

print(kv_kt)

kv_kt.plot(
    kind='bar',
    figsize=(10,6)
)

plt.title('Số lượng thí sinh theo khu vực và khối thi')
plt.xlabel('Khối thi')
plt.ylabel('Số lượng')

plt.grid(axis='y')

plt.show()

# =====================================================
# CÂU 4
# Đậu rớt theo khối thi
# =====================================================

print("===== CÂU 4 =====")

dau_rot_kt = pd.crosstab(
    df['KT'],
    df['KQXT']
)

print(dau_rot_kt)

dau_rot_kt.plot(
    kind='bar',
    figsize=(10,6)
)

plt.title('Đậu rớt theo khối thi')
plt.xlabel('Khối thi')
plt.ylabel('Số lượng')

plt.grid(axis='y')

plt.show()

# =====================================================
# CÂU 5
# Đậu rớt theo khu vực
# =====================================================

print("===== CÂU 5 =====")

dau_rot_kv = pd.crosstab(
    df['KV'],
    df['KQXT']
)

print(dau_rot_kv)

dau_rot_kv.plot(
    kind='bar',
    figsize=(10,6)
)

plt.title('Đậu rớt theo khu vực')
plt.xlabel('Khu vực')
plt.ylabel('Số lượng')

plt.grid(axis='y')

plt.show()

# =====================================================
# CÂU 6
# Đậu rớt theo dân tộc
# =====================================================

print("===== CÂU 6 =====")

dau_rot_dt = pd.crosstab(
    df['DT'],
    df['KQXT']
)

print(dau_rot_dt)

dau_rot_dt.plot(
    kind='bar',
    figsize=(12,6)
)

plt.title('Đậu rớt theo dân tộc')
plt.xlabel('Dân tộc')
plt.ylabel('Số lượng')

plt.grid(axis='y')

plt.show()

# =====================================================
# CÂU 7
# Đậu rớt theo giới tính
# =====================================================

print("===== CÂU 7 =====")

dau_rot_gt = pd.crosstab(
    df['GT'],
    df['KQXT']
)

print(dau_rot_gt)

dau_rot_gt.plot(
    kind='bar',
    figsize=(8,5)
)

plt.title('Đậu rớt theo giới tính')
plt.xlabel('Giới tính')
plt.ylabel('Số lượng')

plt.grid(axis='y')

plt.show()
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from scipy import stats

# =====================================================
# ĐỌC DỮ LIỆU
# =====================================================

url = "https://raw.githubusercontent.com/huynhhoc/DataAnalystDeepLearning/main/Data/dulieuxettuyendaihoc.csv"

df = pd.read_csv(url)

# =====================================================
# TẠO BIẾN PHÂN LỚP T1
# =====================================================

def phan_loai_t1(diem):

    if diem < 5:
        return 'kem'

    elif diem < 7:
        return 'tb'

    elif diem < 8:
        return 'kha'

    else:
        return 'gioi'

df['phanlopT1'] = df['T1'].apply(phan_loai_t1)

# =====================================================
# CÂU 1
# Mô tả và khảo sát phân phối T1
# =====================================================

print("===== CÂU 1 =====")

# -------------------------
# Thống kê mô tả
# -------------------------

print("\nThống kê mô tả T1")

print(df['T1'].describe())

print("\nVariance:", df['T1'].var())

print("Skewness:", df['T1'].skew())

print("Kurtosis:", df['T1'].kurt())

# =====================================================
# BOXPLOT
# =====================================================

plt.figure(figsize=(8,5))

plt.boxplot(df['T1'])

plt.title('Boxplot T1')

plt.ylabel('Điểm T1')

plt.show()

# -------------------------
# 10 đại lượng boxplot
# -------------------------

print("\n===== 10 ĐẠI LƯỢNG =====")

minimum = df['T1'].min()
maximum = df['T1'].max()

q1 = df['T1'].quantile(0.25)
q2 = df['T1'].quantile(0.50)
q3 = df['T1'].quantile(0.75)

iqr = q3 - q1

lower_bound = q1 - 1.5 * iqr
upper_bound = q3 + 1.5 * iqr

mean = df['T1'].mean()
median = df['T1'].median()
std = df['T1'].std()

print("Min:", minimum)
print("Q1:", q1)
print("Median:", q2)
print("Q3:", q3)
print("Max:", maximum)
print("Mean:", mean)
print("Std:", std)
print("IQR:", iqr)
print("Lower Bound:", lower_bound)
print("Upper Bound:", upper_bound)

# =====================================================
# HISTOGRAM
# =====================================================

plt.figure(figsize=(8,5))

plt.hist(df['T1'], bins=10)

plt.title('Histogram T1')

plt.xlabel('T1')
plt.ylabel('Tần số')

plt.grid(True)

plt.show()

# =====================================================
# QQ PLOT
# =====================================================

plt.figure(figsize=(6,6))

stats.probplot(df['T1'], dist="norm", plot=plt)

plt.title('QQ Plot T1')

plt.grid(True)

plt.show()

# =====================================================
# NHẬN XÉT
# =====================================================

print("\n===== NHẬN XÉT =====")

if df['T1'].skew() > 0:
    print("Phân phối lệch phải")
elif df['T1'].skew() < 0:
    print("Phân phối lệch trái")
else:
    print("Phân phối đối xứng")

# =====================================================
# CÂU 2
# Phân phối T1 theo nhóm phanlopT1
# =====================================================

print("\n===== CÂU 2 =====")

# -------------------------
# Boxplot theo nhóm
# -------------------------

plt.figure(figsize=(8,5))

df.boxplot(column='T1', by='phanlopT1')

plt.title('Boxplot T1 theo phân lớp')

plt.suptitle('')

plt.xlabel('Phân lớp')
plt.ylabel('T1')

plt.show()

# -------------------------
# Histogram theo nhóm
# -------------------------

for nhom in df['phanlopT1'].unique():

    plt.figure(figsize=(7,5))

    du_lieu = df[df['phanlopT1'] == nhom]

    plt.hist(du_lieu['T1'], bins=10)

    plt.title(f'Histogram nhóm {nhom}')

    plt.xlabel('T1')
    plt.ylabel('Tần số')

    plt.grid(True)

    plt.show()

# -------------------------
# QQ Plot theo nhóm
# -------------------------

for nhom in df['phanlopT1'].unique():

    plt.figure(figsize=(6,6))

    du_lieu = df[df['phanlopT1'] == nhom]

    stats.probplot(du_lieu['T1'], dist='norm', plot=plt)

    plt.title(f'QQ Plot nhóm {nhom}')

    plt.grid(True)

    plt.show()

# =====================================================
# CÂU 3
# Tương quan DH1 và T1
# =====================================================

print("\n===== CÂU 3 =====")

covariance = df['DH1'].cov(df['T1'])

correlation = df['DH1'].corr(df['T1'])

print("Covariance:", covariance)

print("Correlation:", correlation)

# Scatter Plot

plt.figure(figsize=(8,5))

plt.scatter(df['T1'], df['DH1'])

plt.title('Scatter Plot DH1 và T1')

plt.xlabel('T1')
plt.ylabel('DH1')

plt.grid(True)

plt.show()

# =====================================================
# CÂU 4
# Tương quan DH1 và T1 theo khu vực
# =====================================================

print("\n===== CÂU 4 =====")

for kv in df['KV'].unique():

    du_lieu = df[df['KV'] == kv]

    corr = du_lieu['DH1'].corr(du_lieu['T1'])

    print(f"Khu vực {kv} - Correlation:", corr)

    plt.figure(figsize=(7,5))

    plt.scatter(
        du_lieu['T1'],
        du_lieu['DH1']
    )

    plt.title(f'DH1 và T1 - KV {kv}')

    plt.xlabel('T1')
    plt.ylabel('DH1')

    plt.grid(True)

    plt.show()

# =====================================================
# CÂU 5
# Tương quan DH1 DH2 DH3
# =====================================================

print("\n===== CÂU 5 =====")

# -------------------------
# Ma trận tương quan
# -------------------------

corr_matrix = df[['DH1', 'DH2', 'DH3']].corr()

cov_matrix = df[['DH1', 'DH2', 'DH3']].cov()

print("\nMa trận tương quan")

print(corr_matrix)

print("\nMa trận hiệp phương sai")

print(cov_matrix)

# =====================================================
# Scatter giữa các biến
# =====================================================

# DH1 vs DH2

plt.figure(figsize=(7,5))

plt.scatter(df['DH1'], df['DH2'])

plt.title('DH1 vs DH2')

plt.xlabel('DH1')
plt.ylabel('DH2')

plt.grid(True)

plt.show()

# DH1 vs DH3

plt.figure(figsize=(7,5))

plt.scatter(df['DH1'], df['DH3'])

plt.title('DH1 vs DH3')

plt.xlabel('DH1')
plt.ylabel('DH3')

plt.grid(True)

plt.show()

# DH2 vs DH3

plt.figure(figsize=(7,5))

plt.scatter(df['DH2'], df['DH3'])

plt.title('DH2 vs DH3')

plt.xlabel('DH2')
plt.ylabel('DH3')

plt.grid(True)

plt.show()
```
